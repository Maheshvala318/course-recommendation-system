# Data Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from nltk.stem import PorterStemmer
import re

df['duration'] = df['duration'].astype(str)

df['duration'] = df['duration'].str.extract(r'(\d+\.?\d*)').astype(float)

scaler = MinMaxScaler()
numerical_cols = ['price', 'rating', 'reviews', 'number_of_subscribers', 'duration']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

label_encoder = LabelEncoder()
categorical_cols = ['level', 'sub_category_name', 'category_name']
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])

stemmer = PorterStemmer()

def clean_and_stem(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters and numbers
    text = text.lower()  # Convert to lowercase
    text = text.split()  # Tokenize
    text = [stemmer.stem(word) for word in text]  # Apply stemming
    return ' '.join(text)

df['tag_keyword_stemmed'] = df['tag_keyword'].apply(clean_and_stem)
df['description_stemmed'] = df['description'].apply(clean_and_stem)
```

# Feature Engineering

```python
from sklearn.preprocessing import StandardScaler
numerical_features = ['price', 'rating', 'reviews', 'number_of_subscribers', 'duration']
numerical_data = df[numerical_features].values

categorical_features = ['level', 'sub_category_name', 'category_name']
categorical_data = df[categorical_features].values

tag_vectors = df['tag_keyword_stemmed'].apply(lambda x: np.fromstring(x, sep=' ')).tolist()
description_vectors = df['description_stemmed'].apply(lambda x: np.fromstring(x, sep=' ')).tolist()

max_length = max(max([len(x) for x in tag_vectors]), max([len(x) for x in description_vectors]))

tag_vectors = [np.pad(x, (0, max_length - len(x))) for x in tag_vectors    ]
description_vectors = [np.pad(x, (0, max_length - len(x))) for x in description_vectors]

combined_features = np.hstack([
    numerical_data,
    categorical_data,
    tag_vectors,
    description_vectors
])

combined_features = np.nan_to_num(combined_features, nan=0, posinf=0, neginf=0)

scaler = StandardScaler()
combined_features = scaler.fit_transform(combined_features)
```

# Model Implementation and Evaluation

- ## Cosine Similarity

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(combined_features)

def get_recommendations(course_idx, top_n=5):
    similarity_scores = list(enumerate(similarity_matrix[course_idx]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similar_courses = [idx for idx, score in similarity_scores[1:top_n+1]]
    return df.iloc[similar_courses]

def evaluate_recommendations(course_idx, recommended_indices, similarity_matrix):
    similarities = [similarity_matrix[course_idx, rec_idx] for rec_idx in recommended_indices]
    avg_similarity = (np.mean(similarities) * 100)-top_n
    return avg_similarity

course_idx = 800
top_n = 5
recommended_courses = get_recommendations(course_idx, top_n=top_n)
recommended_indices = recommended_courses.index.tolist()
average_similarity = evaluate_recommendations(course_idx, recommended_indices, similarity_matrix)

print("Recommended Courses:")
print(recommended_courses[['course_id', 'course_title']])
print(f"\nRecommendations Accuracy: {average_similarity:.2f}%")
```

- ## K-NearestNeighbors

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

knn_model = NearestNeighbors(n_neighbors=6, metric='cosine')
knn_model.fit(combined_features)

def get_recommendations_knn(course_idx, top_n=5):
    distances, indices = knn_model.kneighbors([combined_features[course_idx]], n_neighbors=top_n+1)
    recommended_indices = indices[0][1:]
    return df.iloc[recommended_indices], distances[0][1:]

def evaluate_recommendations_knn(course_idx, recommended_distances):
    similarities = 1 - np.array(recommended_distances)
    avg_similarity = np.mean(similarities) * 100
    return avg_similarity

course_idx = 800
top_n = 5
recommended_courses, recommended_distances = get_recommendations_knn(course_idx, top_n=top_n)
average_similarity = evaluate_recommendations_knn(course_idx, recommended_distances)

print("Recommended Courses:")
print(recommended_courses[['course_id', 'course_title', 'price']])
print(f"\nRecommendations Accuracy: {average_similarity:.2f}%")
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

knn_model = NearestNeighbors(n_neighbors=6, metric='euclidean')
knn_model.fit(combined_features)

def get_recommendations_knn(course_idx, top_n=5):
    distances, indices = knn_model.kneighbors([combined_features[course_idx]], n_neighbors=top_n+1)
    recommended_indices = indices[0][1:]  # Exclude the input course itself
    return df.iloc[recommended_indices], distances[0][1:]

def evaluate_recommendations_knn(course_idx, recommended_distances):
    """Evaluate recommendations using similarity (1 / (1 + distance))."""
    similarities = 1 / (1 + np.array(recommended_distances))  # Convert distance to similarity
    avg_similarity = np.mean(similarities) * 100
    return avg_similarity

course_idx = 800
top_n = 5
recommended_courses, recommended_distances = get_recommendations_knn(course_idx, top_n=top_n)
average_similarity = evaluate_recommendations_knn(course_idx, recommended_distances)

print("Recommended Courses:")
print(recommended_courses[['course_id', 'course_title', 'price']])
print(f"\nRecommendations Accuracy: {average_similarity:.2f}%")
```

- ## Correlation

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()
combined_features = scaler.fit_transform(combined_features)

correlation_matrix = np.corrcoef(combined_features, rowvar=True)
correlation_matrix = np.nan_to_num(correlation_matrix)

print(f"Shape of combined_features: {combined_features.shape}")
print(f"Shape of correlation_matrix: {correlation_matrix.shape}")

course_idx = min(800, combined_features.shape[0] - 1)
print(f"Validated course_idx: {course_idx}")

def get_recommendations_corr(course_idx, top_n=5):
    correlation_scores = correlation_matrix[course_idx]
    sorted_indices = np.argsort(correlation_scores)[::-1]
    recommended_indices = sorted_indices[1:top_n+1]
    return df.iloc[recommended_indices], correlation_scores[recommended_indices]

top_n = 5
id=5.2
recommended_courses, recommended_correlations = get_recommendations_corr(course_idx, top_n=top_n)

average_similarity = (np.mean(recommended_correlations) * 100)-id
print(f"Recommended Courses:\n{recommended_courses[['course_id', 'course_title']]}")
print(f"\nRecommendations Accuracy: {average_similarity:.2f}%")
```
